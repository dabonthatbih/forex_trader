```python
def add_indics(df,
               ma=0,
               ema=0,
               momentum=0,
               roc=0,
               atr=0,
               bollinger=0,
               psr=False,
               sok=False,
               sod=0,
               trix=0,
               massi=False,
               vortex=0,
               rsi=0,
               acc_diss=0,
               chaikin=False,
               mfi=0,
               obv=0,
               force=0,
               eom=0,
               cci=0,
               copp=0,
               kelchu=0,
               ultimate_osc=False,
               donchian=0,
               std=0):
    if psr==True:
        df = indicators.ppsr(df)
    if sok==True:
        df = indicators.stochastic_oscillator(df)
    if massi==True:
        df = indicators.mass_index(df)
    if chaikin==True:
        df = indicators.chaikin_oscillator(df)
    if ultimate_osc==True:
        df = indicators.ultimate_oscillator(df)
    if ma != 0:
        for i in ma:
            df = indicators.moving_average(df, i)
    if ema != 0:
        for i in ema:
            df = indicators.exponential_moving_average(df, i)
    if momentum != 0:
        for i in momentum:
            df = indicators.momentum(df, i)
    if roc != 0:
        for i in roc:
            df = indicators.rate_of_change(df, i)
    if atr != 0:
        for i in atr:
            df = indicators.average_true_range(df, i)
    if bollinger != 0:
        for i in bollinger:
            df = indicators.bollinger_bands(df, i)
    if sod != 0:
        for i in sod:
            df = indicators.stochastic_oscillator_d(df, i)
    if trix != 0:
        for i in trix:
            df = indicators.trix(df, i)
    if vortex != 0:
        for i in vortex:
            df = indicators.vortex_indicator(df, i)
    if rsi != 0:
        pass
        """for i in rsi:
            df = indicators.relative_strength_index(df, i)"""
    if acc_diss != 0:
        for i in acc_diss:
            df = indicators.accumulation_distribution(df, i)
    if mfi != 0:
        for i in mfi:
            df = indicators.money_flow_index(df, i)
    if obv != 0:
        for i in obv:
            df = indicators.on_balance_volume(df, i)
    if force != 0:
        for i in force:
            df = indicators.force_index(df, i)
    if eom != 0:
        for i in eom:
            df = indicators.ease_of_movement(df, i)
    if cci != 0:
        for i in cci:
            df = indicators.commodity_channel_index(df, i)
    if copp != 0:
        for i in copp:
            df = indicators.coppock_curve(df, i)
    if kelchu != 0:
        for i in keclhu:
            df = indicators.keltner_channel(df, i)
    if donchian != 0:
        for i in donchian:
            df = indicators.donchian_channel(df, i)
    if std != 0:
        for i in std:
            df = indicators.standard_deviation(df, i)
    return df
```
