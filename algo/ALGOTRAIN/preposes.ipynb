```python
import import_ipynb
import indicators
import add_indicators
import dataset
import matplotlib
```

```python
ma = [10, 21, 50]
ema = [8, 25, 100]
roc = [9, 14, 25]
bollinger = [9, 12]
rsi = [14]
chaikin = True
df = dataset.df
df = add_indicators.add_indics(df, ma = ma, ema = ema, roc = roc, bollinger = bollinger, rsi= rsi)
df.dropna(inplace=True)
df.head(n=20)
```

```python
def predict(close, minim, maxim, future):
    sku = 0
    offset = 0.00100
    offset3 = 0.00100
    if close > (future + offset3) and close > (maxim - offset):
        sku = -1
    if close < (future - offset3) and close < (minim + offset):
        sku = 1
    return sku
```

```python
def future_predict(period):
    df['min'] = df['Low'].iloc[::-1].shift(1).rolling(period).min().iloc[::-1]
    df['max'] = df['High'].iloc[::-1].shift(1).rolling(period).max().iloc[::-1]
    df['future'] = df['Close'].shift(-period)
    df['pred']= list(map(predict, df['Close'], df['min'], df['max'], df['future']))
    df.dropna()
    print(df['pred'].value_counts())
    print(df[['Close', 'Low', 'min', 'max', 'future', 'pred']].head(n=20))
future_predict(6)
```
